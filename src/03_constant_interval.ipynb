# Constant Interval

We will choose a fixed parameter a > 0 for this strategy. In the beginning, we will provide liquidity in a symmetric interval around the current price. This interval will never be adjusted. More specifically, we will choose the interval [p − a, p + a], where p is the price at the beginning.
For the USDC-ETH pool, we will choose a ∈ [10, 40000], which means that the size is between 10 and 40000 basis points, or 0.1% and 400%. It does not make sense to choose another a because, at any time, the price is always in [p−a,p+a].

Import code dependencies

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import date
from decimal import Decimal
from typing import List

from demeter import TokenInfo, Actuator, Strategy, MarketInfo, ChainType
from demeter.uniswap import UniV3Pool, UniLpMarket
from demeter.result import performance_metrics, round_results
from demeter.broker import AccountStatus
```

Set matplotlib to show formatted account status

```python
pd.options.display.max_columns = None
pd.set_option("display.width", 5000)
```

Make plot about price and account value / position net value

```python
# Token and Market setup
usdc = TokenInfo(name="usdc", decimal=6)
eth = TokenInfo(name="eth", decimal=18)
pool_address = "0x88e6a0c2ddd26feeb64f039a2c41296fcb3f5640"
market_key = MarketInfo("market1")
start_date = date(2024, 4, 1)
end_date = date(2024, 9, 30)

# Load market data function
def load_market_data(market, data_path="../data"):
    market.data_path = data_path
    market.load_data(
        chain=ChainType.ethereum.name,
        contract_addr=pool_address,
        start_date=start_date,
        end_date=end_date,
    )
    return market

# Define Constant Interval Strategy
class ConstantIntervalStrategy(Strategy):
    def __init__(self, a=1000):
        super().__init__()
        self.a = a

    def initialize(self):
        market = self.markets[market_key]
        init_price = market.market_status.data.price
        market.even_rebalance(init_price)  # initial rebalance
        
        # Convert `a` to Decimal and calculate bounds
        a_decimal = Decimal(self.a) / Decimal(10000)
        lower_bound = init_price * (1 - a_decimal)
        upper_bound = init_price * (1 + a_decimal)

        # Ensure the lower bound is above zero
        if lower_bound < Decimal("0.01"):
            lower_bound = Decimal("0.01")

        # Add liquidity with corrected bounds
        market.add_liquidity(lower_bound, upper_bound)
        print(f"Initial price: {init_price}, Lower bound: {lower_bound}, Upper bound: {upper_bound}")

# Function to run strategy and capture net value time series
def run_strategy(a_value):
    market = load_market_data(UniLpMarket(market_key, UniV3Pool(usdc, eth, 0.05, usdc)))
    actuator = Actuator()
    actuator.broker.add_market(market)
    actuator.broker.set_balance(usdc, 5000)
    actuator.broker.set_balance(eth, 1)
    actuator.strategy = ConstantIntervalStrategy(a=a_value)
    actuator.set_price(market.get_price_from_data())
    actuator.run()
    
    # Print performance metrics
    metrics = round_results(performance_metrics(
        actuator.account_status_df["net_value"], 
        benchmark=actuator.account_status_df["price"]["ETH"]
    ))
    print(f"\nPerformance for a = {a_value}:", metrics)
    
    # Return net value time series
    return actuator.account_status_df["net_value"]

# Different values of `a` to test
a_values = [10, 50, 100, 500, 1000, 5000, 10000, 20000, 30000, 40000]
results = {}

# Run each strategy and collect net value over time
for a in a_values:
    print(f"\nRunning strategy with a = {a}")
    results[f"a={a}"] = run_strategy(a)

# Convert results to DataFrame for plotting
results_df = pd.DataFrame(results)

# Plot the results with improved aesthetics
plt.figure(figsize=(14, 9))
for a in results_df.columns:
    plt.plot(results_df.index, results_df[a], label=a, linewidth=1.5)
plt.title("Constant Interval Strategy: Net Value Comparison for Different `a` Values", fontsize=16)
plt.xlabel("Date", fontsize=14)
plt.ylabel("Net Value (USDC)", fontsize=14)
plt.legend(loc="upper left", title="`a` Value", fontsize=10)
plt.grid(visible=True, linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()

# Save results to CSV for further analysis
results_df.to_csv("constant_interval_comparison_results.csv")
```
