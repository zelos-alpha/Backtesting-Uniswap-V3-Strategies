# Fill up
This is another strategy trying to solve the problem of 4.4. The first interval is [pc − a, pc + a]. The second interval is [pc, pc + a] or [pc − a, pc], depending on which asset is leftover.

Import code dependencies

```python
from datetime import date, timedelta
from decimal import Decimal
from typing import List, Dict

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

from demeter import (
    TokenInfo,
    Actuator,
    ChainType,
    MarketInfo,
    Strategy,
    PeriodTrigger,
    RowData
)
from demeter.broker import AccountStatus
from demeter.uniswap import UniV3Pool, UniLpMarket
from demeter.result import performance_metrics, round_results
```

```python
from datetime import date, timedelta
from typing import Dict
from _decimal import Decimal
import pandas as pd

from demeter import TokenInfo, Actuator, ChainType, MarketInfo, Strategy, PeriodTrigger, RowData
from demeter.uniswap import UniLpMarket, UniV3Pool
```

Update Interval 2 hours

```python
# Token and Market setup
usdc = TokenInfo(name="usdc", decimal=6)
eth = TokenInfo(name="eth", decimal=18)
pool_address = "0x88e6a0c2ddd26feeb64f039a2c41296fcb3f5640"
pool = UniV3Pool(usdc, eth, 0.05, usdc)
market_key = MarketInfo("market1")
start_date = date(2024, 4, 1)
end_date = date(2024, 9, 30)
market = UniLpMarket(market_key, pool)


# Load market data function
def load_market_data(market, data_path="../data"):
    market.data_path = data_path
    market.load_data(
        chain=ChainType.ethereum.name,
        contract_addr=pool_address,
        start_date=start_date,
        end_date=end_date,
    )
    return market

# Define FillUpStrategy with safeguards for liquidity bounds
class FillUpStrategy(Strategy):
    def __init__(self, a=1000):
        super().__init__()
        self.a = a

    def initialize(self):
        lp_market = self.broker.markets[market_key]
        init_price = lp_market.market_status.data.price

        # Safeguard interval bounds for `a`
        a_decimal = Decimal(self.a) / Decimal(10000)
        lower_bound_a = init_price * (1 - a_decimal)
        upper_bound_a = init_price * (1 + a_decimal)
        if lower_bound_a < Decimal("0.01"):
            lower_bound_a = Decimal("0.01")

        # Add primary and secondary intervals
        lp_market.even_rebalance(init_price)  # initial rebalance
        lp_market.add_liquidity(lower_bound_a, upper_bound_a)
        
        if self.broker.assets[market.base_token].balance > 0:
            lp_market.add_liquidity(lower_bound_a, init_price)
        else:
            lp_market.add_liquidity(init_price, upper_bound_a)
        
        self.triggers.append(PeriodTrigger(time_delta=timedelta(hours=2), do=self.work))
        print(f"Initialized with a = {self.a}, price range: [{lower_bound_a}, {upper_bound_a}]")

    def work(self, row_data: RowData):
        lp_market = self.broker.markets[market_key]
        current_price = row_data.prices[eth.name]

        # Remove and rebalance liquidity
        if len(lp_market.positions) > 0:
            lp_market.remove_all_liquidity()
            lp_market.even_rebalance(current_price)

        # Calculate interval bounds using `a`
        a_decimal = Decimal(self.a) / Decimal(10000)
        lower_bound_a = current_price * (1 - a_decimal)
        upper_bound_a = current_price * (1 + a_decimal)
        if lower_bound_a < Decimal("0.01"):
            lower_bound_a = Decimal("0.01")

        lp_market.add_liquidity(lower_bound_a, upper_bound_a)
        
        if self.broker.assets[market.base_token].balance > 0:
            lp_market.add_liquidity(lower_bound_a, current_price)
        else:
            lp_market.add_liquidity(current_price, upper_bound_a)

        # print(f"Rebalancing at {row_data.timestamp}, a range: [{lower_bound_a}, {upper_bound_a}]")

# Function to run strategy and return net value time series
def run_strategy(a_value):
    market = load_market_data(UniLpMarket(market_key, UniV3Pool(usdc, eth, 0.05, usdc)))
    actuator = Actuator()
    actuator.broker.add_market(market)
    actuator.broker.set_balance(usdc, 5000)
    actuator.broker.set_balance(eth, 1)
    actuator.strategy = FillUpStrategy(a=a_value)
    actuator.set_price(market.get_price_from_data())
    actuator.run()
    
    # Print rounded performance metrics
    metrics = round_results(performance_metrics(
        actuator.account_status_df["net_value"], 
        benchmark=actuator.account_status_df["price"]["ETH"]
    ))
    print(f"\nPerformance for a = {a_value}:", metrics)
    
    # Return net value time series
    return actuator.account_status_df["net_value"]

# Different values of `a` to test
a_values = [10, 50, 100, 500, 1000, 5000, 10000, 20000, 30000, 40000]
results = {}

# Run each strategy and collect net value over time
for a in a_values:
    print(f"\nRunning strategy with a = {a}")
    net_value_series = run_strategy(a)
    results[f"a={a}"] = net_value_series

# Convert results to DataFrame for plotting
results_df = pd.DataFrame(results)

# Plot the results with clear labeling for each value of `a`
plt.figure(figsize=(14, 9))
for label in results_df.columns:
    plt.plot(results_df.index, results_df[label], label=label, linewidth=1.5)
plt.title("Fill Up Strategy: Net Value Comparison for Different `a` Values", fontsize=16)
plt.xlabel("Date", fontsize=14)
plt.ylabel("Net Value (USDC)", fontsize=14)
plt.legend(loc="upper left", title="`a` Values", fontsize=10)
plt.grid(visible=True, linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
