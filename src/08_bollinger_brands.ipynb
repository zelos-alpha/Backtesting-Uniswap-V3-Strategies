# Bollinger Bands
This strategy combines the moving average strategy 4.9 and the volatility sized intervals strategy 4.10. We will provide liquidity inside the Bollinger Bands. These bands are made up of a lower band BOLL = pa −c·v and an upper band BOLU = pa +c·v. The liquidity position will be [pa −v·c,pa +v·c]. This strategy also has an unused leftover token.

```python
import pandas as pd
import matplotlib.pyplot as plt
import math
from datetime import date, timedelta
from demeter import TokenInfo, Actuator, Strategy, RowData, simple_moving_average, ChainType, MarketInfo, PeriodTrigger
from demeter.result import performance_metrics, round_results
from demeter.uniswap import UniV3Pool, UniLpMarket
from demeter.result import performance_metrics, round_results

# Token and Market setup
usdc = TokenInfo(name="usdc", decimal=6)
eth = TokenInfo(name="eth", decimal=18)
pool_address = "0x88e6a0c2ddd26feeb64f039a2c41296fcb3f5640"
market_key = MarketInfo("market1")
start_date = date(2024, 4, 1)
end_date = date(2024, 9, 30)

# Load market data function
def load_market_data(market, data_path="../data"):
    market.data_path = data_path
    market.load_data(
        chain=ChainType.ethereum.name,
        contract_addr=pool_address,
        start_date=start_date,
        end_date=end_date,
    )
    return market

# Define Bollinger Bands Strategy with adjustable c and update_interval
class AddByVolatilityStrategy(Strategy):
    def __init__(self, c=2, update_interval=timedelta(days=1)):
        super().__init__()
        self.c = c
        self.update_interval = update_interval

    def initialize(self):
        self.add_column(market_key, "sma", simple_moving_average(self.data[market_key].price, timedelta(days=1)))
        self.add_column(market_key, "volatility", realized_volatility(self.data[market_key].price, timedelta(days=1), timedelta(days=1)))
        
        # Adjust trigger based on update interval
        self.triggers.append(PeriodTrigger(time_delta=self.update_interval, trigger_immediately=True, do=self.work))
        self.markets.default.even_rebalance(self.data[market_key].price[0])

    def work(self, row_data: RowData):
        lp_market = self.broker.markets[market_key]
        lp_row_data = row_data.market_status[market_key]
        if len(lp_market.positions) > 0:
            lp_market.remove_all_liquidity()
            lp_market.even_rebalance(row_data.prices[eth.name])
        if math.isnan(lp_row_data.volatility):
            return
        limit = self.c * float(row_data.prices[eth.name]) * lp_row_data.volatility
        lp_market.add_liquidity(lp_row_data.sma - limit, lp_row_data.sma + limit)

# Function to run strategy and return net value time series
def run_strategy(c_value, update_interval):
    market = load_market_data(UniLpMarket(market_key, UniV3Pool(usdc, eth, 0.05, usdc)))
    actuator = Actuator()
    actuator.broker.add_market(market)
    actuator.broker.set_balance(usdc, 5000)
    actuator.broker.set_balance(eth, 1)
    actuator.strategy = AddByVolatilityStrategy(c=c_value, update_interval=update_interval)
    actuator.set_price(market.get_price_from_data())
    actuator.run()
    
    # Print rounded performance metrics
    metrics = round_results(performance_metrics(
        actuator.account_status_df["net_value"], 
        benchmark=actuator.account_status_df["price"]["ETH"]
    ))
    print(f"\nPerformance for c = {c_value}, update interval = {update_interval}:", metrics)
    
    # Return net value time series
    return actuator.account_status_df["net_value"]

# Different update intervals to test
update_intervals = [
    timedelta(hours=1), timedelta(hours=2), timedelta(hours=6), timedelta(hours=12), timedelta(hours=18), 
    timedelta(days=1), timedelta(days=3), timedelta(days=7), timedelta(days=14), timedelta(days=30)
]
c_values = [0.5, 1, 2]  # Added c=0.5 for additional test cases
results = {c: {} for c in c_values}  # Separate dictionaries for each c value

# Run each strategy with different update intervals and c values, collect net value over time
for c in c_values:
    for interval in update_intervals:
        print(f"\nRunning strategy with c = {c}, update interval = {interval}")
        net_value_series = run_strategy(c, interval)
        results[c][f"interval={interval}"] = net_value_series

    # Convert results to DataFrame for plotting
    results_df = pd.DataFrame(results[c])

    # Plot the results for the current value of c
    plt.figure(figsize=(14, 9))
    for label in results_df.columns:
        plt.plot(results_df.index, results_df[label], label=label, linewidth=1.5)
    plt.title(f"Bollinger Bands Strategy: Net Value Comparison for Different Intervals (c={c})", fontsize=16)
    plt.xlabel("Date", fontsize=14)
    plt.ylabel("Net Value (USDC)", fontsize=14)
    plt.legend(loc="upper left", title="Update Intervals", fontsize=10)
    plt.grid(visible=True, linestyle='--', linewidth=0.5)
    plt.tight_layout()
    plt.show()
```
