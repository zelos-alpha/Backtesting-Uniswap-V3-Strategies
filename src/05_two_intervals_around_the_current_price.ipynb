# Two Intervals Around the Current Price
This strategy aims to fix the problem of having leftover liquidity. The first interval is the same as in the interval around the current price strategy 4.4, namely [pc − a, pc + a]. Let b be a paramter and b ∈ [10, 1000], then the second interval will either be [pc, pc +b] or [pc −b, pc], depending on which asset is leftover [4]. This will allow us to use up all our liquidity.

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import date, timedelta
from decimal import Decimal
from demeter import TokenInfo, Actuator, ChainType, MarketInfo, Strategy, PeriodTrigger, RowData
from demeter.result import performance_metrics, round_results
from demeter.uniswap import UniV3Pool, UniLpMarket

# Token and Market setup
usdc = TokenInfo(name="usdc", decimal=6)
eth = TokenInfo(name="eth", decimal=18)
pool_address = "0x88e6a0c2ddd26feeb64f039a2c41296fcb3f5640"
market_key = MarketInfo("market1")
pool = UniV3Pool(usdc, eth, 0.05, usdc)
market = UniLpMarket(market_key, pool)
start_date = date(2024, 4, 1)
end_date = date(2024, 9, 30)

# Load market data function
def load_market_data(market, data_path="../data"):
    market.data_path = data_path
    market.load_data(
        chain=ChainType.ethereum.name,
        contract_addr=pool_address,
        start_date=start_date,
        end_date=end_date,
    )
    return market

# Define Two Intervals Around the Current Price Strategy
class TwoIntervalsAroundtheCurrentPriceStrategy(Strategy):
    def __init__(self, a=10, b=10, update_interval=timedelta(hours=2)):
        super().__init__()
        self.a = a
        self.b = b
        self.update_interval = update_interval

    def initialize(self):
        lp_market = self.broker.markets[market_key]
        init_price = lp_market.market_status.data.price

        # Calculate and set bounds with lower bound protection for `a` range
        a_decimal = Decimal(self.a) / Decimal(10000)
        lower_bound_a = init_price * (1 - a_decimal)
        upper_bound_a = init_price * (1 + a_decimal)
        if lower_bound_a < Decimal("0.01"):
            lower_bound_a = Decimal("0.01")
            
        lp_market.add_liquidity(lower_bound_a, upper_bound_a)

        # Calculate bounds for `b` range with similar formula
        b_decimal = Decimal(self.b) / Decimal(10000)
        lower_bound_b = init_price * (1 - b_decimal)
        upper_bound_b = init_price * (1 + b_decimal)
        if lower_bound_b < Decimal("0.01"):
            lower_bound_b = Decimal("0.01")

        # Add additional liquidity for leftover balances using adjusted `b` range
        if self.broker.assets[market.base_token].balance > 0:
            lp_market.add_liquidity(lower_bound_b, init_price)
        else:
            lp_market.add_liquidity(init_price, upper_bound_b)

        # Register trigger to rebalance every `update_interval`
        self.triggers.append(PeriodTrigger(time_delta=self.update_interval, do=self.work))
        print(f"Initial price: {init_price}, a range: [{lower_bound_a}, {upper_bound_a}], b range: [{lower_bound_b}, {upper_bound_b}]")

    def work(self, row_data: RowData):
        lp_market = self.broker.markets[market_key]
        current_price = row_data.prices[eth.name]
        # Remove and rebalance existing liquidity
        if len(lp_market.positions) > 0:
            lp_market.remove_all_liquidity()
            lp_market.even_rebalance(current_price)
        # Calculate interval bounds using `a`
        a_decimal = Decimal(self.a) / Decimal(10000)
        lower_bound_a = current_price * (1 - a_decimal)
        upper_bound_a = current_price * (1 + a_decimal)
        if lower_bound_a < Decimal("0.01"):
            lower_bound_a = Decimal("0.01")

        lp_market.add_liquidity(lower_bound_a, upper_bound_a)

        # Calculate and apply bounds for `b`
        b_decimal = Decimal(self.b) / Decimal(10000)
        if self.broker.assets[market.base_token].balance > 0:
            lower_bound_b = current_price * (1 - b_decimal)
            if lower_bound_b < Decimal("0.01"):
                lower_bound_b = Decimal("0.01")
            lp_market.add_liquidity(lower_bound_b, current_price)
        else:
            upper_bound_b = current_price * (1 + b_decimal)
            lp_market.add_liquidity(current_price, upper_bound_b)

# Function to run strategy and capture net value time series
def run_strategy(a_value, b_value):
    market = load_market_data(UniLpMarket(market_key, UniV3Pool(usdc, eth, 0.05, usdc)))
    actuator = Actuator()
    actuator.broker.add_market(market)
    actuator.broker.set_balance(usdc, 5000)
    actuator.broker.set_balance(eth, 1)
    actuator.strategy = TwoIntervalsAroundtheCurrentPriceStrategy(a=a_value, b=b_value)
    actuator.set_price(market.get_price_from_data())
    actuator.run()
    
    # Print performance metrics
    metrics = round_results(performance_metrics(
        actuator.account_status_df["net_value"], 
        benchmark=actuator.account_status_df["price"]["ETH"]
    ))
    print(f"\nPerformance for a = {a_value}, b = {b_value}:", metrics)
    
    # Return net value time series
    return actuator.account_status_df["net_value"]

# Different values of `a` and `b` to test
a_values = [10, 50, 100, 500, 1000, 5000]
b_values = [50, 100, 500, 1000, 5000, 10000]
results = {}

# Run each combination of `a` and `b` only if a < b
for a in a_values:
    for b in b_values:
        if a < b:
            print(f"\nRunning strategy with a = {a}, b = {b}")
            net_value_series = run_strategy(a, b)
            results[f"a={a}_b={b}"] = net_value_series

# Convert results to DataFrame for plotting
results_df = pd.DataFrame(results)

# Plot the results with clear labeling for each (a, b) combination
plt.figure(figsize=(14, 9))
for label in results_df.columns:
    plt.plot(results_df.index, results_df[label], label=label, linewidth=1.5)
plt.title("Two Intervals Around the Current Price Strategy: Net Value Comparison for Different (a, b) Values", fontsize=16)
plt.xlabel("Date", fontsize=14)
plt.ylabel("Net Value (USDC)", fontsize=14)
plt.legend(loc="upper left", title="(a, b) Values", fontsize=10)
plt.grid(visible=True, linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import date, timedelta
from decimal import Decimal
from demeter import TokenInfo, Actuator, ChainType, MarketInfo, Strategy, PeriodTrigger, RowData
from demeter.result import performance_metrics, round_results
from demeter.uniswap import UniV3Pool, UniLpMarket

# Token and Market setup
usdc = TokenInfo(name="usdc", decimal=6)
eth = TokenInfo(name="eth", decimal=18)
pool_address = "0x88e6a0c2ddd26feeb64f039a2c41296fcb3f5640"
market_key = MarketInfo("market1")
pool = UniV3Pool(usdc, eth, 0.05, usdc)
market = UniLpMarket(market_key, pool)
start_date = date(2024, 4, 1)
end_date = date(2024, 9, 30)

# Load market data function
def load_market_data(market, data_path="../data"):
    market.data_path = data_path
    market.load_data(
        chain=ChainType.ethereum.name,
        contract_addr=pool_address,
        start_date=start_date,
        end_date=end_date,
    )
    return market

# Define Two Intervals Around the Current Price Strategy
class TwoIntervalsAroundtheCurrentPriceStrategy(Strategy):
    def __init__(self, a=10, b=10, update_interval=timedelta(days=14)):
        super().__init__()
        self.a = a
        self.b = b
        self.update_interval = update_interval

    def initialize(self):
        lp_market = self.broker.markets[market_key]
        init_price = lp_market.market_status.data.price

        # Calculate and set bounds with lower bound protection for `a` range
        a_decimal = Decimal(self.a) / Decimal(10000)
        lower_bound_a = init_price * (1 - a_decimal)
        upper_bound_a = init_price * (1 + a_decimal)
        if lower_bound_a < Decimal("0.01"):
            lower_bound_a = Decimal("0.01")
            
        lp_market.add_liquidity(lower_bound_a, upper_bound_a)

        # Calculate bounds for `b` range with similar formula
        b_decimal = Decimal(self.b) / Decimal(10000)
        lower_bound_b = init_price * (1 - b_decimal)
        upper_bound_b = init_price * (1 + b_decimal)
        if lower_bound_b < Decimal("0.01"):
            lower_bound_b = Decimal("0.01")

        # Add additional liquidity for leftover balances using adjusted `b` range
        if self.broker.assets[market.base_token].balance > 0:
            lp_market.add_liquidity(lower_bound_b, init_price)
        else:
            lp_market.add_liquidity(init_price, upper_bound_b)

        # Register trigger to rebalance every `update_interval`
        self.triggers.append(PeriodTrigger(time_delta=self.update_interval, do=self.work))
        print(f"Initial price: {init_price}, a range: [{lower_bound_a}, {upper_bound_a}], b range: [{lower_bound_b}, {upper_bound_b}]")

    def work(self, row_data: RowData):
        lp_market = self.broker.markets[market_key]
        current_price = row_data.prices[eth.name]
        # Remove and rebalance existing liquidity
        if len(lp_market.positions) > 0:
            lp_market.remove_all_liquidity()
            lp_market.even_rebalance(current_price)
        # Calculate interval bounds using `a`
        a_decimal = Decimal(self.a) / Decimal(10000)
        lower_bound_a = current_price * (1 - a_decimal)
        upper_bound_a = current_price * (1 + a_decimal)
        if lower_bound_a < Decimal("0.01"):
            lower_bound_a = Decimal("0.01")

        lp_market.add_liquidity(lower_bound_a, upper_bound_a)

        # Calculate and apply bounds for `b`
        b_decimal = Decimal(self.b) / Decimal(10000)
        if self.broker.assets[market.base_token].balance > 0:
            lower_bound_b = current_price * (1 - b_decimal)
            if lower_bound_b < Decimal("0.01"):
                lower_bound_b = Decimal("0.01")
            lp_market.add_liquidity(lower_bound_b, current_price)
        else:
            upper_bound_b = current_price * (1 + b_decimal)
            lp_market.add_liquidity(current_price, upper_bound_b)

# Function to run strategy and capture net value time series
def run_strategy(a_value, b_value):
    market = load_market_data(UniLpMarket(market_key, UniV3Pool(usdc, eth, 0.05, usdc)))
    actuator = Actuator()
    actuator.broker.add_market(market)
    actuator.broker.set_balance(usdc, 5000)
    actuator.broker.set_balance(eth, 1)
    actuator.strategy = TwoIntervalsAroundtheCurrentPriceStrategy(a=a_value, b=b_value)
    actuator.set_price(market.get_price_from_data())
    actuator.run()
    
    # Print performance metrics
    metrics = round_results(performance_metrics(
        actuator.account_status_df["net_value"], 
        benchmark=actuator.account_status_df["price"]["ETH"]
    ))
    print(f"\nPerformance for a = {a_value}, b = {b_value}:", metrics)
    
    # Return net value time series
    return actuator.account_status_df["net_value"]

# Different values of `a` and `b` to test
a_values = [10, 50, 100, 500, 1000, 5000]
b_values = [50, 100, 500, 1000, 5000, 10000]
results = {}

# Run each combination of `a` and `b` only if a < b
for a in a_values:
    for b in b_values:
        if a < b:
            print(f"\nRunning strategy with a = {a}, b = {b}")
            net_value_series = run_strategy(a, b)
            results[f"a={a}_b={b}"] = net_value_series

# Convert results to DataFrame for plotting
results_df = pd.DataFrame(results)

# Plot the results with clear labeling for each (a, b) combination
plt.figure(figsize=(14, 9))
for label in results_df.columns:
    plt.plot(results_df.index, results_df[label], label=label, linewidth=1.5)
plt.title("Two Intervals Around the Current Price Strategy: Net Value Comparison for Different (a, b) Values", fontsize=16)
plt.xlabel("Date", fontsize=14)
plt.ylabel("Net Value (USDC)", fontsize=14)
plt.legend(loc="upper left", title="(a, b) Values", fontsize=10)
plt.grid(visible=True, linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
